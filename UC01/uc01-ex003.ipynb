```python
import numpy as np
import matplotlib.pyplot as plt
from random import randint

# Criação da classe com valores de referência e funções do programa
class MonitorTemperatura:
  # Estado inicial das variáveis de controle
  def __init__(self):
    self.temperatura_maxima = 80
    self.temperatura_minima = 20
    self.temperatura_anterior = 0 # Primeira variação sempre será 0

  def analisar_temperaturas(self, temperaturas):
    # Calculo da média, desvio, valor máximo e mínimo das temperaturas
    media = np.mean(temperaturas)
    desvio = np.std(temperaturas)
    temperatura_maxima = temperaturas.max()
    temperatura_minima = temperaturas.min()
    print(f"{'Análise dos dados de temperatura':^50}\n")
    print(f"Temperatura média: {media}ºC\nDesvio: {desvio}\nTemperatura máxima: {temperatura_maxima}ºC\nTemperatura mínima: {temperatura_minima}ºC\n")

  def verificar_risco_temperatura(self, temperaturas):
    self.variacao_temperatura = []
    self.temperatura_anterior = temperaturas[0]
    for temp in temperaturas:
      variacao = temp - self.temperatura_anterior
      if temp < self.temperatura_minima:
        print(f"\033[31m⚠️ ATENÇÃO! A temperatura está abaixo de 20ºC!\033[m")
        print(f"\033[37mA temperatura variou {variacao}ºC desde a última análise.\033[m")
        self.temperatura_anterior = temp
        self.variacao_temperatura.append(variacao)

      elif temp > self.temperatura_maxima:
        print(f"\033[31m⚠️ ATENÇÃO! A temperatura está acima de 80ºC!\033[m")
        print(f"\033[37mA temperatura variou {variacao}ºC desde a última análise.\033[m")
        self.temperatura_anterior = temp
        self.variacao_temperatura.append(variacao)

      else:
        print(f"\033[32m✔️ A temperatura está dentro do intervalo seguro.\033[m")
        print(f"\033[37mA temperatura variou {variacao}ºC desde a última análise.\033[m")
        self.temperatura_anterior = temp
        self.variacao_temperatura.append(variacao)

      print(f"\033[37m🌡️ Temperatura atual: {temp}ºC\033[m")
      print("-"*50)
    self.analisar_temperaturas(temperaturas)
    self.gerar_estatisticas(temperaturas)

  def gerar_estatisticas(self, temperaturas):
    # Criação de gráficos estatísticos utilizando MatPlotLib
    # Gráfico da temperatura x tempo
    plt.figure(figsize=(6, 4))
    plt.plot(temperaturas, marker="o", color="blue")
    plt.title("Temperatura x Tempo")
    plt.xlabel("Tempo")
    plt.ylabel("Temperatura (ºC)")
    plt.grid(True)

    # Gráfico da variação de temperatura x tempo
    plt.figure(figsize=(6, 4))
    plt.plot(self.variacao_temperatura, marker="o", color="red")
    plt.title("Variação x Tempo")
    plt.xlabel("Tempo")
    plt.ylabel("Variação de Temperatura (ºC)")
    plt.grid(True)

    plt.show()

# Randomiza 10 valores entre 0 e 150
dados = [randint(0, 150) for x in range(10)]

# Criação de um array unidimensional com NumPy
temperatura = np.array(dados)

# Faz o chamado da função que analisa os valores da temperatura
monitor = MonitorTemperatura()
monitor.verificar_risco_temperatura(temperatura)
```
