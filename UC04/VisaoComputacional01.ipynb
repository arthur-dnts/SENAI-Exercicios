```python
# Importações de bibliotecas
import cv2
from matplotlib import pyplot as plt
```

```python
# Carregamento do arquivo de imagem
image = cv2.imread("test.jpg")

# Converte a imagem (colorida) para escala de cinza
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Exibição da imagem em RGB (conversão BGR -> RGB) via matplotlib
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
# Detecção de rostos
image = cv2.imread("test.jpg")

# Carregando o classificador Cascade pré-treinado
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

# Convertendo a imagem a escala de cinza
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Executando a detecção de rostos, retornando uma lista da posição relativa (x, y, largura e altura)
detected_faces = face_cascade.detectMultiScale(image_gray, scaleFactor=1.3, minNeighbors=5)

# Verificando cada face detectada
for (start_x, start_y, width, height) in detected_faces:
  # Calculando o ponto final (canto inferior direito) do retângulo
  end_x = start_x + width
  end_y = start_y + height

  # Desenha o retângulo azul ao redor do rosto detectado
  cv2.rectangle(
      img = image,
      pt1 = (start_x, start_y),
      pt2 = (end_x, end_y),
      color = [255, 0, 0],
      thickness = 2
  )

# Convertendo a imagem com as detecções de BGR para RGB e exibindo-a
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("Rosotos detectados (em azul)")
plt.axis("off") # Remove os eixos para deixar a exibição limpa
plt.show()
```
