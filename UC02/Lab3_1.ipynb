```python
import warnings
import requests
import zipfile
import io
import pandas as pd
from scipy.io import arff

warnings.simplefilter("ignore")

# Requisição à URL do dataset UCI Vertebral Column
zip_file = "https://archive.ics.uci.edu/static/public/212/vertebral+column.zip"

r = requests.get(zip_file, stream=True) # Stream p/ maiores arquivos

# Unzip do arquivo baixado
vertebral_zip = zipfile.ZipFile(io.BytesIO(r.content))
vertebral_zip.extractall()

# Lendo o arquivo ARFF
data = arff.loadarff("column_2C_weka.arff")

# Criando um data frame pandas com os dados carregados e exibindo os 5 primeiros itens
df = pd.DataFrame(data[0])

df.head()
```
