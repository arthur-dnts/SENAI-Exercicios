```python
# Importação de bibliotecas e datasets
import pandas as pd
import requests
import zipfile
import io

pd.set_option("display.max_rows", 500)
pd.set_option("display.max_columns", 500)
pd.set_option("display.width", 1000)

# Requisição para download do dataset e extração do arquivo
zip_file = "https://archive.ics.uci.edu/static/public/10/automobile.zip"

r = requests.get(zip_file, stream=True)

automobile = zipfile.ZipFile(io.BytesIO(r.content))
automobile.extractall()
```

```python
# Manipulação dos dados
data = "automobile/imports-85.data"

col_names=[
    "symboling", "normalized-losses", "fuel-type", "aspiration",
    "num-of-doors", "body-style", "drive-wheels", "engine-location",
    "wheel-base", "length", "width", "height", "curb-weight", "engine-type",
    "num-of-cylinders", "engine-size", "fuel-system", "bore", "stroke",
    "compression-ratio", "horsepower", "peak-rpm", "city-mpg", "highway-mpg", "price"
]

# Leitura do arquivo CSV e criação do data frame
df_car = pd.read_csv(data, sep=",", names=col_names, na_values="?", header=None)
```

```python
# Verificando o shape e imprimindo os primeiros registros

df_car.shape
df_car.head(5)
```

```python
# Exibindo as informações
df_car.info()
```

```python
# Verificando as colunas
df_car.columns
```

```python
# Copiando colunas
df_car = df_car[["aspiration", "num-of-doors", "drive-wheels", "num-of-cylinders"]].copy()
```

```python
# Examinando os dados
df_car.head()
```

```python
# Codificação de componentes
df_car.info()
```

```python
# Determinando os valores ordinais
df_car["num-of-doors"].value_counts()
```

```python
# Mapeamento com dicionário
door_mapper = {
    "two": 2,
    "four": 4
}
```

```python
# Gerando uma nova coluna com base no dicionário criado
df_car["doors"] = df_car["num-of-doors"].replace(door_mapper)
```

```python
# Exibindo o data frame
df_car.head()
```

```python
# Gerando novas colunas
df_car["num-of-cylinders"].value_counts()
```

```python
# Criando mapeador
cylinder_mapper = {
    "two": 2,
    "three": 3,
    "four": 4,
    "five": 5,
    "six": 6,
    "eight": 8,
    "twelve": 12
}
```

```python
# Mapear com replace
df_car["cylinders"] = df_car["num-of-cylinders"].replace(cylinder_mapper)
```

```python
# Exibindo o DataFrame
df_car.head()
```

```python
# Codificando os dados categóricos não ordinais
df_car["drive-wheels"].value_counts()
```

```python
# Adicionando novos componentes
df_car = pd.get_dummies(df_car, columns=["drive-wheels"])
```

```python
# Exiibindo componentes
df_car.head()
```

```python
# Calculando e exibindo a contagem de frequência
df_car["aspiration"].value_counts()
```

```python
# Aplicando o One-Hot Encoder
df_car = pd.get_dummies(
    df_car,
    columns=["aspiration"],
    drop_first=True
)
```
