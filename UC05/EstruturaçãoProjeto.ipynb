## 📃Contexto
> As instituições e empresas estão cada vez mais interessadas em soluções que utilizem Inteligência Artificial para automatizar processos e melhorar a experiência das pessoas. Uma dessas aplicações é o reconhecimento facial, que pode ser usado em diferentes contextos, como:

- **Segurança**, para controlar o acesso de pessoas autorizadas em ambientes restritos;

- **Educação**, para realizar a chamada automática de alunos em sala de aula;

- **Comércio**, para identificar o perfil dos clientes e personalizar o atendimento;

- **Saúde**, para monitorar pacientes e detectar expressões faciais associadas a dor ou desconforto.

## 🎯 Objetivo
> Planejar o desenvolvimento de um sistema de Reconhecimento Facial Simples que faz a leitura de uma imagem e faça a validação como autorizado ou não autorizado, aplicando o raciocínio de um projeto de Inteligência Artificial desde a definição do problema à escolha da técnica.

---

Nas empresas, o controle de acesso de colaboradores e visitantes ainda é feito de forma manual ou com cartões físicos, o que gera filas, riscos de extravio e falhas de segurança. Um sistema de reconhecimento facial pode automatizar o processo de identificação, reduzindo erros e aumentando a eficiência operacional.

Desenvolver um sistema de reconhecimento facial para controle de acesso empresarial, capaz de identificar automaticamente funcionários e visitantes autorizados, garantindo agilidade, rastreabilidade e segurança na entrada das instalações.
